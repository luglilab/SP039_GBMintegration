```python
import seaborn
sns.set(rc={'font.family':'Arial', 'font.serif':'Arial'})
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(20, 8))
ax = sns.boxplot(data=df2, x="louvain_sub_renamed", y="0",hue="Type",hue_order=['PrimaryTumor','Recurrent'],
                 whis=[0, 100], width=.4,dodge=True)
ax = sns.stripplot(x="louvain_sub_renamed", y="0", data=df2,hue="Type",
                   dodge=True,hue_order=['PrimaryTumor','Recurrent'],legend=False,
              size=3, color=".3", linewidth=1)


annotator = Annotator(ax, box_pairs, data=df2, x="louvain_sub_renamed", y="0",hue="Type", hue_order=['PrimaryTumor','Recurrent'])

annotator.configure(test='t-test_welch', text_format='star', loc='inside',comparisons_correction=None)
annotator.apply_and_annotate()

ax.xaxis.grid(False)
ax.set(ylabel="Relative Fraction")
ax.set(xlabel="Cluster")

sns.despine(trim=True, left=False)
plt.xticks(rotation=45)
plt.tight_layout()
f.tight_layout()
f.savefig('/home/lugli/spuccio/Projects/SP039/GMB_revisionSept2023/figures/Cluster_fraction_with0.pdf',
            dpi=f.dpi,bbox_inches='tight',format="pdf")
```
